```python
import pandas as pd 
```

```python
data = pd.read_csv('D:\SEKOLAH\KULIAH\SEMESTER 6\PENAMBANGAN DATA\Clustering K-Means\customer_data_6.csv')
```

```python
new_data = data
```

```python
for x in new_data.index:
  if new_data.loc[x, "MaritalStatus"] == 'belum menikah':
    new_data.loc[x, "MaritalStatus"] = 0
  else : 
     new_data.loc[x, "MaritalStatus"] = 1 
```

```python
new_data.head()
```

```python
for x in new_data.index:
  if new_data.loc[x, "Gender"] == 'laki-laki':
    new_data.loc[x, "Gender"] = 0
  else : 
     new_data.loc[x, "Gender"] = 1 
```

```python
new_data.head()
```

```python
for x in new_data.index:
  if new_data.loc[x, "HomeOwner"] == 'tidak memiliki rumah sendiri':
    new_data.loc[x, "HomeOwner"] = 0
  else : 
     new_data.loc[x, "HomeOwner"] = 1 
```

```python
new_data.head()
```

```python
new_data['Education'].tolist()
```

```python
for x in new_data.index:
  if new_data.loc[x, "Education"] == 'SMA':
    new_data.loc[x, "Education"] = 0
  elif new_data.loc[x,"Education"] == 'S1' : 
    new_data.loc[x, "Education"] = 1
  elif new_data.loc[x,"Education"] == 'Post-graduate' : 
    new_data.loc[x, "Education"] = 2
  elif new_data.loc[x,"Education"] == 'S2' : 
    new_data.loc[x, "Education"] = 3
  elif new_data.loc[x,"Education"] == 'S3' : 
    new_data.loc[x, "Education"] = 4
```

```python
new_data
```

```python
new_data['Education'].value_counts()
```

```python
for x in new_data.index:
  if new_data.loc[x, "Occupation"] == 'sales':
    new_data.loc[x, "Occupation"] = 0
  elif new_data.loc[x,"Occupation"] == 'skilled manual work' : 
    new_data.loc[x, "Occupation"] = 1
  elif new_data.loc[x,"Occupation"] == 'unskilled manual work' : 
    new_data.loc[x, "Occupation"] = 2
  elif new_data.loc[x,"Occupation"] == 'professional' : 
    new_data.loc[x, "Occupation"] = 3
  elif new_data.loc[x,"Occupation"] == 'clerical work' : 
    new_data.loc[x, "Occupation"] = 4
  elif new_data.loc[x,"Occupation"] == 'management' : 
    new_data.loc[x, "Occupation"] = 5
```
